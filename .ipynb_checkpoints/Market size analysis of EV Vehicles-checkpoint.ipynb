```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("Electric_Vehicle_Population_Data.csv")
data.head()
```

### Details about data:- 
- **VIN (1-10):** Partial Vehicle Identification Number.
- **County:** The county in which the vehicle is registered.
- **City:** The city in which the vehicle is registered.
- **State:** The state in which the vehicle is registered. 
- **Postal Code:** The postal code where the vehicle is registered.
- **Model Year:** The year of the vehicle model.
- **Make:** The manufacturer of the vehicle.
- **Model:** The model of the vehicle.
- **Electric Vehicle Type:** The type of electric vehicle, e.g., Battery Electric Vehicle (BEV).
- **Clean Alternative Fuel Vehicle (CAFV) Eligibility:** Eligibility status for clean alternative fuel vehicle programs.
- **Electric Range:** The maximum range of the vehicle on a single charge (in miles).
- **Base MSRP:** The Manufacturer’s Suggested Retail Price.
- **Legislative District:** The legislative district where the vehicle is registered.
- **DOL Vehicle ID:** Department of Licensing Vehicle Identification.
- **Vehicle Location:** Geographic coordinates of the vehicle location.
- **Electric Utility:** The electric utility service provider for the vehicle’s location.
- **2020 Census Tract:** The census tract for the vehicle’s location.

```python
# EV Adoption over time
ev_per_year=data["Model Year"].value_counts().sort_index()

sns.set_style("whitegrid")
plt.figure(figsize=(12,6),dpi=100)
sns.barplot(x=ev_per_year.index,y=ev_per_year.values,palette="viridis")
plt.title("EV Adoption Over Time")
plt.xlabel("Adoption Year")
plt.ylabel("Number of vehicles registered")
plt.tight_layout()
plt.show()
```

## Observations:-
- From the above bar chart, it’s clear that EV adoption has been increasing over time, especially noting a __significant upward trend starting around 2016.__ The number of vehicles registered grows modestly up until that point and then begins to __rise more rapidly from 2017 onwards.__ The year __2023 shows a particularly sharp increase__ in the number of registered EVs, with the bar for 2023 being the highest on the graph, __indicating a peak in EV adoption.__

```python
# EV Adoption country wise
ev_country=data["County"].value_counts()

sns.set_style("whitegrid")
plt.figure(figsize=(12,6),dpi=100)
sns.barplot(x=ev_country.index,y=ev_country.values,palette="viridis")
plt.title("EV Adotion based on Country")
plt.xlabel("Country")
plt.ylabel("Number of vehicles registered")
plt.xticks(rotation=90)
plt.show()
```

## Observations:-
- From that graph it is clear that EV adotion is more in __King,Snohomish,Pierce,Clark.__

```python
# Getting top 4 countries as a list
top_country=ev_country.head(4).index

# Filtering the data regrarding to top 4 countries
top_country_ev=data[data["County"].isin(top_country)]
```

```python
# Grouping the data as per county wise and city wise
top_city=top_country_ev.groupby(["County","City"]).agg({"VIN (1-10)":"count"}).reset_index()
top_city=top_city.rename(columns={"VIN (1-10)":"Number of cars"})
top_city=top_city.sort_values(by=["Number of cars"],ascending=[False])
```

```python
# Ordering the data in descending order and in country wise
top_city_ev=top_city.sort_values(by=["County","Number of cars"],ascending=[True,False])
```

```python
# Visualization of top 10 cities over all

sns.set_style("whitegrid")
plt.figure(figsize=(12,6),dpi=100)
sns.barplot(x='Number of cars', y='City', hue='County', data=top_city.head(10), palette="magma",dodge=False)
plt.title("EV Adotion of top 10 cities")
plt.ylabel("City")
plt.xlabel("Number of vehicles registered")
plt.show()

```

## Observations:
- The above graph compares the number of electric vehicles registered in various cities within three counties: King, Clark, Snohomish, and Pierce. The horizontal bars represent cities, and their length corresponds to the number of vehicles registered, colour-coded by county. Here are the key findings from the above graph:

## <center>Estimated Market Size Analysis Of Electric Vehicles</center>

```python
# Number of EVs Registered each year

ev_count=data["Model Year"].value_counts().sort_index()
ev_count
```

## Observations:
- The dataset provides the number of electric vehicles registered each year from 1997 through 2024. However, the data for 2024 is incomplete as it only contains the data till March. Here’s a summary of EV registrations for recent years:
    - In 2021, there were 19,063 EVs registered.
    - In 2022, the number increased to 27708 EVs.
    - In 2023, a significant jump to 57,519 EVs was observed.
    - For 2024, currently, 7,072 EVs are registered, which suggests partial data.

- To forecast the total number of EVs expected to be registered in 2024, we can use a growth rate based approach from previous complete years.



```python
# Forecasting EV
from scipy.optimize import curve_fit
import numpy as np

# Filtering data set for the complete years only
complete_years=ev_count[ev_count.index<=2023]

# function for exponential growth
def expo_growth(x,a,b):
    return a*np.exp(b*x)

# Prepare data for curve fitting
x_data=complete_years.index-complete_years.index.min() # Normalization for numerical stability while fitting
y_data=complete_years.values

# Fitting the data for Exponential growth function
params, covariance = curve_fit(expo_growth,x_data,y_data)

# use params to forecast the number of EVs for 2024 to 2030
forecast_years=np.arange(2024,2024+7) - complete_years.index.min()
forecasted_value=expo_growth(forecast_years,*params)
forecasted_years=forecast_years+complete_years.index.min()
rounded_forecasted_values=np.round(forecasted_value).astype(int)

# Creating Data Frame for Forecasted Values
forecasted_ev=pd.DataFrame({"year":forecasted_years,
                           "EVs":rounded_forecasted_values})
forecasted_ev
```

```python
# Visualization of Current and Forecasted EV Market


plt.figure(figsize=(14, 8), dpi=100)

# Plot the actual registrations
plt.plot(complete_years.index, complete_years.values, "bo-", label="Actual Registrations", markersize=8)

# Plot the forecasted registrations
plt.plot(forecasted_ev["year"], forecasted_ev["EVs"], "ro--", label="Forecasted Registrations", markersize=8)

plt.title("Current and Estimated EV Market", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Number of EV Registrations", fontsize=14)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend(fontsize=12)

# Highlight key points
max_actual = complete_years.values.max()
max_forecast = forecasted_ev["EVs"].max()
plt.annotate(f'Max Actual\n{max_actual}', xy=(complete_years.index[-1], max_actual), 
             xytext=(complete_years.index[-1]-2, max_actual+10000),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=12, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'))
plt.annotate(f'Max Forecast\n{max_forecast}', xy=(forecasted_ev["year"].iloc[-1], max_forecast), 
             xytext=(forecasted_ev["year"].iloc[-1]-2, max_forecast+10000),
             arrowprops=dict(facecolor='red', shrink=0.05),
             fontsize=12, bbox=dict(facecolor='white', edgecolor='red', boxstyle='round,pad=0.5'))

# Highlight forecast period
plt.axvspan(forecasted_ev["year"].min(), forecasted_ev["year"].max(), color='red', alpha=0.1)

plt.show()

```

## Observations:
- The number of actual EV registrations remained relatively low and stable until around 2010, after which there was a consistent and steep upward trend, suggesting a significant increase in EV adoption.
- The forecasted EV registrations predict an even more dramatic increase in the near future, with the number of registrations expected to rise sharply in the coming years.

__Given the growing trend in actual EV registrations and the projected acceleration as per the forecast data, we can conclude that the EV market size is expected to expand considerably. The steep increase in forecasted registrations suggests that consumer adoption of EVs is on the rise, and this trend is likely to continue. Overall, the data point towards a promising future for the EV industry, indicating a significant shift in consumer preferences and a potential increase in related investment and business opportunities.__
